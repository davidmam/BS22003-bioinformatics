```python
fh=open('link_emboss_e.txt')
ofh=open('enzymes.txt','w')
for line in fh.readlines():
    if '#' not in line:
        fields=line.split()
        #print(fields)
        if fields[1].upper().count('A')+ fields[1].upper().count('C') +fields[1].upper().count('G')+ fields[1].upper().count('T')==len(fields[1]):
            if fields[3] == '2' and int(fields[2]) >3:
                ofh.write('\t'.join(fields[:1]+[fields[1].upper()]+fields[5:7])+"\n")
ofh.close()
                
```
